```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV files
baseline = pd.read_csv('E:///Depth4UNet/train_history copy/metrics.csv')               # baseline (RGB only?)
unet     = pd.read_csv('E:///Depth4UNet/train_history copy/Unet_dice_by_epoch.csv')    # UNet (RGB)
depth    = pd.read_csv('E:///Depth4UNet/train_history copy/depth_dice_by_epoch.csv')   # UNet + Depth (RGBD)
GELU    = pd.read_csv('E:///Depth4UNet/train_history copy/depth_dice_by_epoch.csv')   # UNet + Depth (RGBD)
# Create the plot
plt.figure(figsize=(8,5))
plt.plot(baseline['epoch'], baseline['dice'], label='UNet_VGG11 (metrics.csv)')
plt.plot(unet['epoch'], unet['dice'], label='UNet (Unet_dice_by_epoch.csv)')
plt.plot(depth['epoch'], depth['dice'], label='Depth4UNet (depth_dice_by_epoch.csv)')
plt.plot(GELU['epoch'], GELU['dice'], label='GELU (GELU.csv)')
plt.xlabel('Epoch')
plt.ylabel('Dice Score')
plt.title('Dice Score vs. Epoch for Three Models')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load PR curve data
baseline_pr = pd.read_csv('E:///Depth4UNet/train_history copy/pr_curve.csv')           # baseline (metrics.csv counterpart)
unet_pr     = pd.read_csv('E:///Depth4UNet/train_history copy/Unet_pr_curve.csv')      # UNet (RGB)
depth_pr    = pd.read_csv('E:///Depth4UNet/train_history copy/depth_pr_curve.csv')     # UNet + Depth (RGBD)
GELU    = pd.read_csv('E:\\\Depth4UNet\train_history copy\pr_curve(gelu).csv')   # UNet + Depth (RGBD)
# Plot PR curves
plt.figure(figsize=(7,6))

plt.plot(baseline_pr['recall'], baseline_pr['precision'], label='UNet (RGB)', linewidth=2)
plt.plot(unet_pr['recall'],     unet_pr['precision'],     label='UNet_VG11 (RGB)',   linewidth=2)
plt.plot(depth_pr['recall'],    depth_pr['precision'],    label='Depth4UNet (RGBD)', linewidth=2)
plt.plot(GELU['recall'],    GELU['precision'],    label='GELU (RGBD)', linewidth=2)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('PrecisionRecall Curves for Three Models')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
